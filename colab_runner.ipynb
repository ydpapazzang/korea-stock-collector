## 4️⃣ 결과 확인

```python
import glob
import pandas as pd

# 최신 파일 자동 선택
files = sorted(glob.glob('data/korea_stocks_*.csv'))
if not files:
    print('❌ 수집된 파일이 없습니다.')
else:
    latest = files[-1]
    df = pd.read_csv(latest)
    print(f'📁 파일: {latest}')
    print(f'📊 종목수: {len(df):,}개')
    print(f'📋 컬럼: {list(df.columns)}')
    print()
    display(df.head(20))
```

```python
# 정배열 종목만 필터링
golden = df[df['GoldenAlign_YN'] == 'Y'].copy()
print(f'✨ 정배열 종목: {len(golden):,}개')
display(golden.sort_values('MarketCap', ascending=False).head(20))
```
